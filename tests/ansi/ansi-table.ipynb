```python
# You must install the 'rich' library to run this script:
# pip install rich

# See https://github.com/knuesel/callisto/issues/17, thanks @syqwq-OMG

from rich.console import Console
from rich.table import Table
from rich import box

def print_model_summary():
    """
    Creates and prints a colored model summary table,
    similar to a Keras model output, using the 'rich' library.
    """
    
    # Data from the image
    # (Layer Name, Layer Type, Output Shape, Params)
    table_data = [
        ("dense_20", "Dense", "(None, 128)", "3,456"),
        ("dropout_10", "Dropout", "(None, 128)", "0"),
        ("dense_21", "Dense", "(None, 64)", "8,256"),
        ("dropout_11", "Dropout", "(None, 64)", "0"),
        ("dense_22", "Dense", "(None, 32)", "2,080"),
        ("dense_23", "Dense", "(None, 1)", "33"),
    ]

    # Initialize a console object
    console = Console()

    # Create a table instance
    # We use 'box.HEAVY_HEAD' to get a bold line under the header
    table = Table(
        title="Model Summary",
        box=box.HEAVY_HEAD,
        show_header=True,
        header_style="bold white",
        show_lines=True,
        border_style="white"
    )

    # Add columns to the table
    # We right-justify the 'Param #' column
    table.add_column("Layer (type)", style="white", no_wrap=True, width=30)
    table.add_column("Output Shape", style="white")
    table.add_column("Param #", style="white", justify="right")

    # Add rows to the table using 'rich' markup for colors
    for layer, type, shape, params in table_data:
        
        # [bright_blue]...[/bright_blue] makes the text light blue
        layer_str = f"{layer} ([bright_blue]{type}[/bright_blue])"
        
        shape_str = f"[red]{shape}[/red]"
        
        params_str = f"[green]{params}[/green]"
        
        table.add_row(layer_str, shape_str, params_str)

    # Print the table to the console
    console.print(table)

if __name__ == "__main__":
    print_model_summary()
```
